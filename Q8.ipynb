Ω = {BB, BG, GB, GG}
Each of these four outcomes has a probability of $1/4$

C: {BG, GB, GG}
(B-atleast one child is a girl)

A ={GG}

P(A ¦ C) = $\frac{P(A
\cap C)}{P(C)}$

(probability that both children are girls, given that at least one of them is a girl )

```python
import numpy as np

#1.Total Sample Space (Ω)
#where 0 = Boy, 1 = Girl
num_samples = 1000000
child1 = np.random.randint(0,2,num_samples)
child2 = np.random.randint(0,2,num_samples)

#2.Condition (Event C):"At least one child is a girl"
at_least_one_girl = (child1 == 1) | (child2 == 1)
s_samples_c1 = child1[at_least_one_girl]
s_samples_c2 = child2[at_least_one_girl]

#3.Target (Event A):"Both children are girls"
both_girls = (s_samples_c1 == 1) & (s_samples_c2 == 1)

# 4. Final Answer
probability = np.mean(both_girls)
print(f"Probability of both being girls: {probability:.4f}")
```
