```python
import math

a = 8   # Minimum
b = 20  # Maximum
c = 11  # Most Likely

Ex = (a+4 * c+b) / 6
variance = ((Ex-a)*(b-Ex)) / 7
std_deviation = math.sqrt(variance)

print(f"6.Expectation E[X]:      {Ex:.2f} days")
print(f"7.Variance Var(X):       {variance:.4f}")
print(f"8.Standard Deviation:    {std_deviation:.4f} days")
```
