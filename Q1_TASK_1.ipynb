```python
import numpy as np
import plotly.graph_objects as go

theta = np.linspace(-4, 4, 200)
fig = go.Figure()

for b in [-1, 0, 1]:
    p = 1 / (1 + np.exp(-1.5 * (theta - b)))
    fig.add_trace(go.Scatter(x=theta, y=p, name=f"a=1.5, b={b}"))


p_low_a = 1 / (1 + np.exp(-0.5 * (theta - 0)))
fig.add_trace(go.Scatter(x=theta, y=p_low_a, name="a=0.5, b=0"))

fig.update_layout(title="2PL IRT Curves", xaxis_title="Ability (theta)", yaxis_title="P(Y=1)")
fig.show()
```

Increase of the difficulty parameter $b_i$ shifts the logistic curve horizontally to the right. Mathematically, it requires a higher latent ability level $\theta$ to maintain a 50% probability of a correct response ($p_i(b_i) = 0.5$).
