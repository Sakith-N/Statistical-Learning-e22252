```python
import numpy as np
import pandas as pd

#G1:M1 Good, G2:M2 Good, G3:M1 Defective, G4:M2 Defective
omega = ['G1','G2','G3','G4']
p_measure = [0.5,0.2,0.1,0.2]

#X: Quality (0=Good,1=Defective)
#Y: Detailed Outcome(1,2,3,4)
df = pd.DataFrame({
    'Outcome': omega,
    'P': p_measure,
    'X': [0, 0, 1, 1],
    'Y': [1, 2, 3, 4]
})

#1.probability space
print(f"1.Omega:{omega}")
print(f"  P: {p_measure}")

#compute the sigma of X
print(f"2.sigma(X) is the set of events generated by Quality: {{ {{w1,w2}}, {{w3,w4}} }}")
#compute the sigma of Y
print(f"3.sigma(Y) is the set of events generated by Details: {{ {{w1}}, {{w2}}, {{w3}}, {{w4}} }}")

#4.Comparison
print(f"4.sigma(X) subset of sigma(Y) is True. Y contains more information.")

#5 & 6.Marginal Distributions
marg_x = df.groupby('X')['P'].sum()
marg_y = df.groupby('Y')['P'].sum()
print(f"5. Marginal X (Quality): P(0)={marg_x[0]:.1f}, P(1)={marg_x[1]:.1f}")
print(f"6. Marginal Y (Outcome): {marg_y.to_dict()}")

#7.Joint Probabilities
print(f"7. P(X=0, Y=1) = {df.iloc[0]['P']}")
print(f"   P(X=1, Y=3) = {df.iloc[2]['P']}")

#8.Conditional Probabilities
p_x1_y3 = df.loc[df['Y']==3, 'P'].values[0] / marg_y[3]
p_y3_x1 = df.loc[df['Y']==3, 'P'].values[0] / marg_x[1]
print(f"8. P(X=1 | Y=3) = {p_x1_y3:.1f}")
print(f"   P(Y=3 | X=1) = {p_y3_x1:.4f} (1/3)")

#9 & 10.Interpretation
print(f"9.Knowing Y removes uncertainty about X (P=1.0).")
print(f"10.Info in Y not in X: The specific Machine ID.")

#11 & 12.Measurability and Conclusion
print(f"11.X is a function of Y, thus it is sigma(Y)-measurable.")
print(f"12.Finer sigma-algebras (like sigma(Y)) allow for more precise predictions.")
```
