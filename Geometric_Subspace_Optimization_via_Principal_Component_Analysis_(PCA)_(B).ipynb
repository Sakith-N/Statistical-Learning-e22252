Fundamental Equation of FA ProofGiven $Z_i = \Lambda F_i + \epsilon_i$, we expand the correlation matrix $R$:$$R = \mathbb{E}[Z_i Z_i^T] = \mathbb{E}\left[(\Lambda F_i + \epsilon_i)(\Lambda F_i + \epsilon_i)^T\right]$$$$= \mathbb{E}\left[\Lambda F_i F_i^T \Lambda^T + \epsilon_i \epsilon_i^T + \Lambda F_i \epsilon_i^T + \epsilon_i F_i^T \Lambda^T\right]$$Pulling deterministic loading structures out of the expectations:$$R = \Lambda \mathbb{E}[F_i F_i^T] \Lambda^T + \mathbb{E}[\epsilon_i \epsilon_i^T] + \Lambda \mathbb{E}[F_i \epsilon_i^T] + \mathbb{E}[\epsilon_i F_i^T] \Lambda^T$$Applying model constraints ($\mathbb{E}[F_i F_i^T] = I_k$, $\mathbb{E}[\epsilon_i \epsilon_i^T] = \Psi$, and cross-orthogonality $\mathbb{E}[\epsilon_i F_i^T] = \mathbf{0}$):$$R = \Lambda I_k \Lambda^T + \Psi + \mathbf{0} + \mathbf{0} = \Lambda \Lambda^T + \Psi$$

2. A raw PCA vector maximizes global variance sequentially, thi leads to crossloadings where a single sensor loads across multiple components. A Varimax rotated factor loading vector optimizes a clean orthogonal layout, steering coefficients toward extreme values to align each sensor with exactly one distinct physical factor.

Let $T$ be an orthogonal rotation matrix ($T T^T = I$). The rotated loading matrix is $\Lambda^* = \Lambda T$. Substituting this back into the covariance model:$$\Lambda^* (\Lambda^*)^T = (\Lambda T)(\Lambda T)^T = \Lambda \left(T T^T\right) \Lambda^T = \Lambda I \Lambda^T = \Lambda \Lambda^T$$Because the underlying matrix product remains unchanged, the global matrix approximation and individual sensor communalities ($h_j^2$, the row sums of squares) are unaffected by the rotation.

3. We construct the joint partitioned vector $Y_i = [Z_i^T, F_i^T]^T$. Its cross-covariance parameters are:$$\mathbb{E}[Z_i Z_i^T] = R, \quad \mathbb{E}[F_i F_i^T] = I_k$$$$\mathbb{E}[Z_i F_i^T] = \mathbb{E}[(\Lambda F_i + \epsilon_i) F_i^T] = \Lambda \mathbb{E}[F_i F_i^T] + \mathbb{E}[\epsilon_i F_i^T] = \Lambda I_k + \mathbf{0} = \Lambda$$This yields the complete partitioned matrix block layout:$$\Sigma_{YY} = \begin{bmatrix} R & \Lambda \\ \Lambda^T & I_k \end{bmatrix}$$Using the conditional expectation projection identity for multivariate normal distributions:
$\mathbb{E}[X_2 | x_1] = \mu_2 + \Sigma_{21}\Sigma_{11}^{-1}(x_1 - \mu_1)$$Mapping $X_2 \to F_i$ and $x_1 \to z_i$ (where $\mu_F = \mathbf{0}, \mu_Z = \mathbf{0}$):$$\Sigma_{11} = R, \quad \Sigma_{21} = \Lambda^T$
