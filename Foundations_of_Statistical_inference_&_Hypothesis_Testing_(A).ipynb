The raw Maximum Likelihood Estimator (MLE) for the population covariance matrix is defined as:$\hat{\Sigma}_{MLE} = \frac{1}{n}\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T$ To evaluate its expectation, we can decompose the deviation from the sample mean by referencing the true population mean $\mu$:$$X_i - \hat{\mu}_n = (X_i - \mu) - (\hat{\mu}_n - \mu)$$Expanding the outer product summation:$$\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T = \sum_{i=1}^{n} \left[ (X_i - \mu) - (\hat{\mu}_n - \mu) \right] \left[ (X_i - \mu) - (\hat{\mu}_n - \mu) \right]^T$$$$= \sum_{i=1}^{n}(X_i - \mu)(X_i - \mu)^T - \sum_{i=1}^{n}(X_i - \mu)(\hat{\mu}_n - \mu)^T - \sum_{i=1}^{n}(\hat{\mu}_n - \mu)(X_i - \mu)^T + \sum_{i=1}^{n}(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$$Since $\sum_{i=1}^{n}(X_i - \mu) = n(\hat{\mu}_n - \mu)$, the middle terms consolidate with the last term:$$\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T = \sum_{i=1}^{n}(X_i - \mu)(X_i - \mu)^T - n(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$$Now, we apply the expectation operator $\mathbb{E}[\cdot]$ to both sides. By definition, since the $X_i$ variables are independent and identically distributed (i.i.d.):$$\mathbb{E}[(X_i - \mu)(X_i - \mu)^T] = \Sigma$$$$\mathbb{E}[(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T] = \text{Var}(\hat{\mu}_n) = \frac{1}{n}\Sigma$$Substituting these back into the expectation of our summation yields:$$\mathbb{E}\left[\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T\right] = \sum_{i=1}^{n}\Sigma - n\left(\frac{1}{n}\Sigma\right) = n\Sigma - \Sigma = (n-1)\Sigma$$Thus, taking the expectation of the original $\hat{\Sigma}_{MLE}$ structure:$$\mathbb{E}[\hat{\Sigma}_{MLE}] = \mathbb{E}\left[\frac{1}{n}\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T\right] = \frac{n-1}{n}\Sigma$$Because $\mathbb{E}[\hat{\Sigma}_{MLE}] \neq \Sigma$
the raw MLE is downwardly biased in finite samples. To eliminate this bias and get the unique unbiased estimator $S$, we apply Bessel's correction, multiplying the expression by $\frac{n}{n-1}$:$$S = \frac{n}{n-1}\hat{\Sigma}_{MLE} = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T$$

$\mathbb{E}[S] = \Sigma$

*   Type I Error ($\alpha$): The probability of rejecting the null hypothesis ($H_0$) when it is actually true.this is a false alarm concluding that a healthy asset is damaged or failing.
*   Type II Error ($\beta$): The probability of failing to reject the null hypothesis ($H_0$) when it is false. In an SHM context, this is a missed detection failing to flag a real fault.

Mathematical Consequences of Setting α=0.0001:

On Statistical Power ($1-\beta$): By shrinking $\alpha$ to an ultra conservative limit,decision boundary would shift into the tails of the null distribution. While this virtually eliminates false alarms, it dramatically expands the acceptance region for $H_0$.This will increases Type II error rate ($\beta$) , severely lowering the statistical power ($1-\beta$). The diagnostic layer becomes blind to low magnitude or early stage structural anomalies.

On the Geometric Volume of the Ellipsoid: The healthy operation confidence boundary is defined by a Mahalanobis distance threshold corresponding to the critical value of a $\chi^2$ distribution. Lowering $\alpha$ directly raises this critical value, geometrically scaling up the volume of the confidence ellipsoid. This wide, loose boundary swallows up structural drift patterns, treating them as nominal variations.
