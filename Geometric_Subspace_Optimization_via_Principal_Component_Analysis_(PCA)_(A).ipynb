1. $\tilde{X}_i = X_i - \mu$, $\Sigma = \mathbb{E}[\tilde{X}_i \tilde{X}_i^T] = P \Lambda P^T$,
$Z_i = P^T \tilde{X}_i$:$$\mathbb{E}[Z_i Z_i^T] = \mathbb{E}[(P^T \tilde{X}_i)(P^T \tilde{X}_i)^T] = \mathbb{E}[P^T \tilde{X}_i \tilde{X}_i^T P]$$Since the projection eigenvector matrix $P$ is a deterministic matrix derived from the archive, it pulls out of the expectation:$$\mathbb{E}[Z_i Z_i^T] = P^T \mathbb{E}[\tilde{X}_i \tilde{X}_i^T] P = P^T \Sigma P$$Substituting the spectral decomposition of $\Sigma$:$$\mathbb{E}[Z_i Z_i^T] = P^T \left(P \Lambda P^T\right) P = \left(P^T P\right) \Lambda \left(P^T P\right)$$Because $P$ is an orthogonal transformation matrix, $P^T P = I_m$. Thus:$$\mathbb{E}[Z_i Z_i^T] = I_m \Lambda I_m = \Lambda$$

Because $\Lambda$ is strictly diagonal, the cross-covariances between distinct coordinates $Z_{i,j}$ and $Z_{i,k}$ ($j \neq k$) drop to zero. Statistically, this means the linear correlations between your sensor channels have been completely decoupled. Geometrically, PCA performs a rigid rotation of your system axes, re-aligning the coordinate framework with the principal directions of data variance.
