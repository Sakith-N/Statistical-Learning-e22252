1. $Z(T)=\sum_{j=0}^{\infty}e^{-E_{j}/(k_{B}T)}$

Let $\beta = \frac{1}{k_B T}$. Substituting the energy eigenvalues $E_j$:


$Z(T) = \sum_{j=0}^{\infty} e^{-\beta \hbar\omega (j + 1/2)} = e^{-\frac{\beta\hbar\omega}{2}} \sum_{j=0}^{\infty} \left(e^{-\beta\hbar\omega}\right)^j$


Since $T > 0$, the expression $e^{-\beta\hbar\omega} < 1$, forming a convergent infinite geometric series with a first term of $1$ and a common ratio of $e^{-\beta\hbar\omega}$. Using the sum formula $S = \frac{a}{1-r}$:


$Z(T) = e^{-\frac{\beta\hbar\omega}{2}} \cdot \frac{1}{1 - e^{-\beta\hbar\omega}} = \frac{e^{-\frac{\hbar\omega}{2k_{B}T}}}{1-e^{-\frac{\hbar\omega}{k_{B}T}}}$

2.

 Helmholtz Free Energy(F):

  $F(T) = -k_B T \ln Z = -k_B T \ln \left( \frac{e^{-\frac{\beta\hbar\omega}{2}}}{1 - e^{-\beta\hbar\omega}} \right)$

  $F(T) = -k_B T \left[ -\frac{\hbar\omega}{2k_B T} - \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) \right] = \frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)$
  

Internal Energy(U):

  $U(T) = \frac{\hbar\omega}{2} + \frac{\hbar\omega e^{-\beta\hbar\omega}}{1 - e^{-\beta\hbar\omega}} = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\frac{\hbar\omega}{k_B T}} - 1}$





3.
Using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$:

$S(T) = -\frac{\partial}{\partial T}\left[ \frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) \right]$

$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) - k_B T \cdot \frac{1}{1 - e^{-\frac{\hbar\omega}{k_B T}}} \cdot \left(-e^{-\frac{\hbar\omega}{k_B T}}\right) \cdot \left(\frac{\hbar\omega}{k_B T^2}\right)$

$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) + \frac{\hbar\omega}{T\left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)}$


Verification of U = F + TS:

$F + TS = \left[\frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)\right] + T \left[-k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) + \frac{\hbar\omega}{T\left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)}\right]$


The terms containing $\ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)$ cancel out perfectly, yielding:


$F + TS = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\frac{\hbar\omega}{k_B T}} - 1} = U(T) $


4. $C_V = \frac{\partial U}{\partial T} = \hbar\omega \cdot \frac{\partial}{\partial T}\left( e^{\frac{\hbar\omega}{k_B T}} - 1 \right)^{-1}$


$C_V = \hbar\omega \cdot (-1)\left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)^{-2} \cdot e^{\frac{\hbar\omega}{k_B T}} \cdot \left(-\frac{\hbar\omega}{k_B T^2}\right) = k_B \left(\frac{\hbar\omega}{k_B T}\right)^2 \frac{e^{\frac{\hbar\omega}{k_B T}}}{\left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)^2}$

5. in the high-temperature limit where $k_B T \gg \hbar\omega$, the variable $x = \frac{\hbar\omega}{2k_B T} \to 0$. Using the small-angle Taylor expansion approximation $\sinh(x) \approx x$:

$C_V \approx k_B x^2 \cdot \frac{1}{x^2} = k_B$
