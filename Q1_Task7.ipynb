```python
import numpy as np
import plotly.graph_objects as go

np.random.seed(42)
theta_true = 0.75
n_items = 20
grid = np.linspace(-4, 4, 1000)
delta_theta = grid[1] - grid[0]

posterior = np.exp(-grid**2 / 2) / np.sqrt(2 * np.pi)

b_arr = np.random.normal(0, 1, n_items)
a_arr = np.random.uniform(0.5, 2.0, n_items)
theta_mean_track = []
theta_map_track = []

for k in range(n_items):
    p_true = 1 / (1 + np.exp(-a_arr[k] * (theta_true - b_arr[k])))
    yk = 1 if np.random.uniform(0, 1) < p_true else 0

    p_grid = 1 / (1 + np.exp(-a_arr[k] * (grid - b_arr[k])))
    likelihood = (p_grid ** yk) * ((1 - p_grid) ** (1 - yk))

    posterior *= likelihood
    posterior /= np.sum(posterior) * delta_theta

    theta_mean_track.append(np.sum(grid * posterior) * delta_theta)
    theta_map_track.append(grid[np.argmax(posterior)])

fig = go.Figure()
fig.add_trace(go.Scatter(y=theta_mean_track, mode='lines+markers', name='Posterior Mean'))
fig.add_trace(go.Scatter(y=theta_map_track, mode='lines+markers', name='MAP Estimate'))
fig.add_trace(go.Scatter(x=[0, n_items-1], y=[theta_true, theta_true], mode='lines', name='True Theta', line=dict(dash='dash', color='red')))
fig.update_layout(title="Sequential Ability Tracking", xaxis_title="Item Number (k)", yaxis_title="Estimate")
fig.show()
```
