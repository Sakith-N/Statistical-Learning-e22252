$\mathbb{E}[x_{k}^{-}] = \mathbb{E}[A_{k-1}x_{k-1}^{+} + G_{k-1}w_{k-1}] = A_{k-1}\mathbb{E}[x_{k-1}^{+}] + G_{k-1}\mathbb{E}[w_{k-1}]$


Since $\mathbb{E}[x_{k-1}^{+}] = m_{k-1}$ and $\mathbb{E}[w_{k-1}] = 0$:

$m_{k}^{-} = A_{k-1}m_{k-1}$

Therefore:

$P_{k}^{-} = A_{k-1}\mathbb{E}[(x_{k-1}^{+} - m_{k-1})(x_{k-1}^{+} - m_{k-1})^T]A_{k-1}^T + G_{k-1}\mathbb{E}[w_{k-1}w_{k-1}^T]G_{k-1}^T + \text{cross-terms}$

Because $x_{k-1}^{+}$ and $w_{k-1}$ are mutually independent, their cross-covariance is zero.
 Knowing $\mathbb{E}[(x_{k-1}^{+} - m_{k-1})(x_{k-1}^{+} - m_{k-1})^T] = P_{k-1}$ and $\mathbb{E}[w_{k-1}w_{k-1}^T] = \Sigma_p$:

$P_{k}^{-} = A_{k-1}P_{k-1}A_{k-1}^{T}+G_{k-1}\Sigma_{p}G_{k-1}^{T}$

The mean of $y_{k}^{-}$ is:

$\mathbb{E}[y_{k}^{-}] = \mathbb{E}[H_{k}x_{k}^{-} + z_{k}] = H_{k}\mathbb{E}[x_{k}^{-}] + \mathbb{E}[z_{k}] = H_{k}m_{k}^{-} + 0 = H_{k}m_{k}^{-}$

The covariance of $y_{k}^{-}$ is:

$\text{Var}(y_{k}^{-}) = \mathbb{E}\left[(y_{k}^{-} - \mathbb{E}[y_{k}^{-}])(y_{k}^{-} - \mathbb{E}[y_{k}^{-}])^T\right]$

Substitute $y_{k}^{-} - \mathbb{E}[y_{k}^{-}] = H_{k}(x_{k}^{-} - m_{k}^{-}) + z_{k}$:

$\text{Var}(y_{k}^{-}) = \mathbb{E}\left[\left(H_{k}(x_{k}^{-} - m_{k}^{-}) + z_{k}\right)\left(H_{k}(x_{k}^{-} - m_{k}^{-}) + z_{k}\right)^T\right]$

Since $x_{k}^{-}$ and $z_k$ are independent, the cross-terms vanish:


$\text{Var}(y_{k}^{-}) = H_{k}\mathbb{E}\left[(x_{k}^{-} - m_{k}^{-})(x_{k}^{-} - m_{k}^{-})^T\right]H_{k}^T + \mathbb{E}[z_{k}z_{k}^T] = H_{k}P_{k}^{-}H_{k}^{T}+\Sigma_{m}$

$\text{Cov}(x_{k}^{-}, y_{k}^{-}) = \mathbb{E}\left[(x_{k}^{-} - m_{k}^{-})(y_{k}^{-} - \mathbb{E}[y_{k}^{-}])^T\right]$

BY Substitute $y_{k}^{-} - \mathbb{E}[y_{k}^{-}] = H_{k}(x_{k}^{-} - m_{k}^{-}) + z_{k}$:


$\text{Cov}(x_{k}^{-}, y_{k}^{-}) = \mathbb{E}\left[(x_{k}^{-} - m_{k}^{-})\left(H_{k}(x_{k}^{-} - m_{k}^{-}) + z_{k}\right)^T\right]$

Since $x_{k}^{-}$ and $z_k$ are independent, $\mathbb{E}\left[(x_{k}^{-} - m_{k}^{-})z_{k}^T\right] = 0$. Therefore:

$\text{Cov}(x_{k}^{-}, y_{k}^{-}) = P_{k}^{-}H_{k}^T$

$\text{Cov}(y_{k}^{-}, x_{k}^{-}) = \left(\text{Cov}(x_{k}^{-}, y_{k}^{-})\right)^T = H_{k}P_{k}^{-}$

$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left(\begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}\right)$

Then the conditional distribution of $X$ given $Y = y$ is Gaussian with:

$\mu_{X|Y} = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$

$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$


Let the Kalman Gain be defined as $K_k = \Sigma_{XY}\Sigma_{YY}^{-1}$:

$K_{k} = P_{k}^{-}H_{k}^{T}(H_{k}P_{k}^{-}H_{k}^{T}+\Sigma_{m})^{-1}$

Thus, the conditional mean is:

$$m_{k} = m_{k}^{-} + K_{k}(y_{k}^{obs} - H_{k}m_{k}^{-})$$

And the conditional covariance is:

$P_{k} = P_{k}^{-} - K_{k}(H_{k}P_{k}^{-}) = (I - K_{k}H_{k})P_{k}^{-}$
